# 2 — Refresh Daily Sales

Reads from the `transactions` Delta table, aggregates by date / store / payment type, and writes to the `daily_sales` Delta table.

**Prerequisite:** Run notebook 1 first so transactions data exists.

```python
# ── Date Range Config ──────────────────────────────────────────────

from datetime import datetime

today = datetime.now()
DATE_FROM = today.strftime("%Y-%m-%d")  # e.g. "2025-03-01"
DATE_TO   = today.strftime("%Y-%m-%d")  # e.g. "2025-03-31"

print(f"Filtering transactions: {DATE_FROM} → {DATE_TO}")
```

```python
# ── Load Transactions & Build Daily Sales ──────────────────────────

from pipeline.transforms import load_transactions, build_daily_sales

txn = load_transactions(date_from=DATE_FROM, date_to=DATE_TO)
print(f"Loaded {len(txn):,} transaction rows")

daily_df = build_daily_sales(txn)
print(f"Generated {len(daily_df):,} daily sales rows")
daily_df.head(10)
```

```python
# ── Write to Delta Table ──────────────────────────────────────────

from pipeline.transforms import save_delta, DAILY_SALES_SCHEMA
from pipeline.config import DAILY_SALES_TABLE

save_delta(daily_df, DAILY_SALES_TABLE, DAILY_SALES_SCHEMA, DATE_FROM, DATE_TO)
print("Done.")
```
