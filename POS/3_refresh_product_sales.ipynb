# 3 — Refresh Product Sales

Reads from the `transactions` Delta table, aggregates by date / store / product, and writes to the `product_sales` Delta table.

**Prerequisite:** Run notebook 1 first so transactions data exists.

```python
# ── Date Range Config ──────────────────────────────────────────────

from datetime import datetime

today = datetime.now()
DATE_FROM = today.strftime("%Y-%m-%d")  # e.g. "2025-03-01"
DATE_TO   = today.strftime("%Y-%m-%d")  # e.g. "2025-03-31"

print(f"Filtering transactions: {DATE_FROM} → {DATE_TO}")
```

```python
# ── Load Transactions & Build Product Sales ───────────────────────

from pipeline.transforms import load_transactions, build_product_sales

txn = load_transactions(date_from=DATE_FROM, date_to=DATE_TO)
print(f"Loaded {len(txn):,} transaction rows")

product_df = build_product_sales(txn)
print(f"Generated {len(product_df):,} product sales rows")
product_df.head(10)
```

```python
# ── Write to Delta Table ──────────────────────────────────────────

from pipeline.transforms import save_delta, PRODUCT_SALES_SCHEMA
from pipeline.config import PRODUCT_SALES_TABLE

save_delta(product_df, PRODUCT_SALES_TABLE, PRODUCT_SALES_SCHEMA, DATE_FROM, DATE_TO)
print("Done.")
```
